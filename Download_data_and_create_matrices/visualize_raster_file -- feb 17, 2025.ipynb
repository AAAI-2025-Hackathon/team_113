```python
import rasterio
from rasterio.plot import show
from rasterio.plot import show_hist

# Path to your GeoTIFF file
tif_path = "USGS_13_n43w124_20241001.tif"

# Open the GeoTIFF file
with rasterio.open(tif_path) as src:
    # Print metadata about the file
    print("File Metadata:")
    print(src.meta)
    
    # Read the first band (assuming single-band data)
    band1 = src.read(1)
    
    # Display the bounds and CRS (Coordinate Reference System)
    print("Bounds:", src.bounds)
    print("CRS:", src.crs)
    
    # Show the raster
    show(band1, title="Raster Visualization")
    
    # Plot a histogram of the values
    show_hist(band1, bins=50, lw=0.0, stacked=False, alpha=0.3, title="Histogram")

# At this point, you can process `band1`, which is a 2D NumPy array.
```

```python
import rasterio
import numpy as np

# Path to your GeoTIFF file
tif_path = "USGS_13_n43w124_20241001.tif"

# Open the GeoTIFF file
with rasterio.open(tif_path) as src:
    # Read all bands as a matrix (array)
    raster_matrix = src.read()  # This returns a 3D matrix: (bands, rows, cols)
    
    # If you only have a single band, you can access it as follows:
    # For single-band raster, it'll be a 2D array
    if raster_matrix.shape[0] == 1:
        raster_matrix = raster_matrix[0, :, :]  # Extract the first (and only) band

    # Print the shape of the matrix
    print("Raster matrix shape:", raster_matrix.shape)

    # Optionally, print the raster metadata
    print("File Metadata:")
    print(src.meta)
    
    # Display the bounds and CRS (Coordinate Reference System)
    print("Bounds:", src.bounds)
    print("CRS:", src.crs)
```

```python
# Save the raster matrix to a .npy file
np.save('elevation_matrix.npy', raster_matrix)
```

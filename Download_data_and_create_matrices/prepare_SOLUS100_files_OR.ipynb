```python
import rasterio
import numpy as np
from pyproj import Proj, transform

# Path to your GeoTIFF file
tif_path = "SOLUS_100_raster_files/anylithicdpt_cm_p.tif"

# Latitude and longitude bounds (replace these with your actual bounds)
max_lat = 43.00055555579519  # minimum latitude
min_lat = 41.9994444436071  # maximum latitude
min_lon = -124.00055555649317  # minimum longitude
max_lon = -122.99944444340576  # maximum longitude

# Open the GeoTIFF file
with rasterio.open(tif_path) as src:
    # Print metadata about the file
    print("File Metadata:")
    print(src.meta)
    
    # Read the first band (assuming single-band data)
    band1 = src.read(1)
    
    # Display the bounds and CRS (Coordinate Reference System)
    print("Bounds:", src.bounds)
    print("CRS:", src.crs)
    
    # Define the source CRS (WGS84, EPSG:4326) for lat/lon
    src_crs = Proj(init='epsg:4326')  # WGS84 (lat/lon)
    target_crs = Proj(init='epsg:5070')  # Raster's CRS (EPSG:5070)
    
    # Transform latitude and longitude (min/max) from EPSG:4326 to EPSG:5070
    min_x, min_y = transform(src_crs, target_crs, min_lon, max_lat)  # Top-left corner in target CRS
    max_x, max_y = transform(src_crs, target_crs, max_lon, min_lat)  # Bottom-right corner in target CRS
    
    # Print transformed coordinates for debugging
    print(f"Transformed coordinates:")
    print(f"Top-left corner (min_x, min_y): ({min_x}, {min_y})")
    print(f"Bottom-right corner (max_x, max_y): ({max_x}, {max_y})")
    
    # Check if the transformed coordinates are within the raster's bounds
    if not (src.bounds.left <= min_x <= src.bounds.right and src.bounds.top >= min_y >= src.bounds.bottom):
        print(f"Warning: Transformed coordinates are outside raster bounds. Please check your lat/lon bounds.")
    
    # Convert the transformed coordinates back to pixel coordinates in the raster
    min_row, min_col = src.index(min_x, min_y)  # Convert to pixel (row, col)
    max_row, max_col = src.index(max_x, max_y)  # Convert to pixel (row, col)
    
    # Print pixel coordinates for debugging
    print(f"Pixel coordinates (min_row, min_col): ({min_row}, {min_col})")
    print(f"Pixel coordinates (max_row, max_col): ({max_row}, {max_col})")
    
    # Check if the pixel coordinates are within the raster's bounds
    if min_row < 0 or min_col < 0 or max_row > src.height or max_col > src.width:
        print(f"Error: The pixel coordinates are outside the raster's bounds.")
    else:
        # Extract the sub-matrix from the raster
        cropped_band = band1[min_row:max_row, min_col:max_col]
        np.save('AAAI_hackathon/anylithicdpt_2D_matrix.npy', cropped_band)
```

```python
import rasterio
import numpy as np
from pyproj import Proj, transform

# Latitude and longitude bounds (replace these with your actual bounds)
max_lat = 43.00055555579519  # minimum latitude
min_lat = 41.9994444436071  # maximum latitude
min_lon = -124.00055555649317  # minimum longitude
max_lon = -122.99944444340576  # maximum longitude

variable = ['caco3','cec7','claytotal','dbovendry','ec','ecec','fragvol','gypsum','ph1to1h2o','sandco','sandfine','sandmed',
    'sandtotal','sandvc','sandvf','sar','silttotal','soc']
depth = ['0','5','15','30','60','100','150']
for v in variable:
    matrix = []
    for d in depth:
        tif_path = f"SOLUS_100_raster_files/{v}_{d}_cm_p.tif"
        with rasterio.open(tif_path) as src:
            # Read the first band (assuming single-band data)
            band1 = src.read(1)
            src_crs = Proj(init='epsg:4326')  # WGS84 (lat/lon)
            target_crs = Proj(init='epsg:5070')  # Raster's CRS (EPSG:5070)

            # Transform latitude and longitude (min/max) from EPSG:4326 to EPSG:5070
            min_x, min_y = transform(src_crs, target_crs, min_lon, max_lat)  # Top-left corner in target CRS
            max_x, max_y = transform(src_crs, target_crs, max_lon, min_lat)  # Bottom-right corner 
            min_row, min_col = src.index(min_x, min_y)  # Convert to pixel (row, col)
            max_row, max_col = src.index(max_x, max_y)  # Convert to pixel (row, col)
            # Extract the sub-matrix from the raster
            cropped_band = band1[min_row:max_row, min_col:max_col]
            matrix.append(cropped_band)
    matrix = np.stack(matrix, axis=0)
    print(v, " : ",matrix.shape)
    np.save(f'AAAI_hackathon/{v}_matrix.npy', matrix)
    del matrix, cropped_band, band1
```
